### 보완 전 100점 코드

```python
def solution2(moves):
  row_dic = {}
  col_dic = {}

  m = [0,0]
  for move in moves:
    m_1 = []
    if move == "U" or move =="D": # 만약 움직임이 "U"(up)이나 "D"(down)이다.(그럼 행의 값 고정)
      x=m[0]
      m_1.append(x)
      if move =="U": 
        m_1.append(m[1] + 1)
        b = m + m_1
      else:
        m_1.append(m[1] - 1)
        b = m_1 + m
      if 6 not in b and -6 not in b: # 6이 넘어가면 안돼
        m = m_1
        if str(b) not in row_dic:
          row_dic[str(b)] = 1
    else:
      y = m[1]    
      if move =="L":
        m_1.append(m[0]-1)
        m_1.append(y)
        b = m + m_1
      else:
        m_1.append(m[0]+1)
        m_1.append(y)
        b = m_1 + m
      if 6 not in b and -6 not in b: # 6이 넘어가면 안돼
        m = m_1
        if str(b) not in col_dic:
          col_dic[str(b)] = 1
  return len(row_dic)+len(col_dic)
```

```python
# 입출력 예
print(solution2("ULURRDLLU"))  # 7
print(solution2("LULLLLLLU"))  # 7
print(solution2("LRLRL")) #1
print(solution2("UDU")) # 1
print(solution2("RRRRRRRRRRRRRRRRRRRRRUUUUUUUUUUUUULU")) #11
print(solution2("LURDLURDLURDLURDRULD")) # 7
print(solution2("UUUUDUDUDUUU")) # 5
print(solution2("LLLLRLRLRLL")) # 5
```

### 프로그래머스 100점 후 보완 코드

```python
def solution(moves):
  x,y = 0,0
  mov = {"U":(0,1),"D":(0,-1),"R":(1,0) ,"L":(-1,0)}
  move_dic = {}
	
  for move in moves:
    dx,dy = mov[move]
    x1,y1 = x+dx,y+dy
    if move=="U" or move=="R":
	    l = [x,y,x1,y1] 
    else:
      l = [x1,y1,x,y]
    if 6 not in l and -6 not in l:
      x,y = x1,y1
      if str(l) not in move_dic:
        move_dic[str(l)] = 1

  return len(move_dic)
```

```python
# 입출력 예
print(solution("ULURRDLLU"))  # 7
print(solution("LULLLLLLU"))  # 7
print(solution("LRLRL")) #1
print(solution("UDU")) #1
print(solution("RRRRRRRRRRRRRRRRRRRRRUUUUUUUUUUUUULU")) #11
print(solution("LURDLURDLURDLURDRULD")) # 7
print(solution("UUUUDUDUDUUU")) # 5
print(solution("LLLLRLRLRLL")) # 5
```
